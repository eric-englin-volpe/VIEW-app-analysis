```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
import pandas as pd
```

```python
df = pd.read_excel("Brian_Eric_entries.xlsx")
```

```python
kenworth = df.loc[df['Model']=="T800"]
bmw = df.loc[df['Model']=="320i"]
dennis = df.loc[df['Model']=="Elite NAS1"]

kenworth_eric = kenworth.loc[kenworth['Who Entered?']=="Eric"]
bmw_eric = bmw.loc[bmw['Who Entered?']=="Eric"]
dennis_eric = dennis.loc[dennis['Who Entered?']=="Eric"]

kenworth_brian = kenworth.loc[kenworth['Who Entered?']=="Brian"]
bmw_brian = bmw.loc[bmw['Who Entered?']=="Brian"]
dennis_brian = dennis.loc[dennis['Who Entered?']=="Brian"]
```

```python
columns = ['Percent Visible Volume',
       'Percent Visible Volume in Front',
       'Percent Visible Volume in Passenger Side', 'Total Front Volume',
       'Total Passenger Volume', 'Total A Pillar Volume','Preschool Children in Blindzone',
       'Grade School Child Bicyclists in Blindzone',
       'Grade School Children in Blindzone', 'Wheelchair users in Blindzone',
       'Bicyclists in Blindzone', 'Adults in Blindzone']
```

```python
for x in columns:
    print(x)
    kwargs = dict(alpha=0.5, bins=5)

    plt.hist(kenworth_brian[x], **kwargs, color='honeydew', hatch = "/", edgecolor='black',label='Brian - Kenworth')
    plt.hist(kenworth_eric[x], **kwargs, color='forestgreen', edgecolor='black',label='Eric - Kenworth')
    plt.hist(bmw_brian[x], **kwargs, color='lightsteelblue', hatch = "/", edgecolor='black',label='Brian Drawings - BMW')
    plt.hist(bmw_eric[x], **kwargs, color='midnightblue', edgecolor='black',label='Eric - BMW')
    plt.hist(dennis_brian[x], **kwargs, color='lightcoral', hatch = "/", edgecolor='black',label='Brian - Dennis Eagle')
    plt.hist(dennis_eric[x], **kwargs, color='darkred', edgecolor='black',label='Eric - Dennis Eagle')
    
    
    plt.gca().set(title=x, ylabel='Frequency')
    plt.legend();
    plt.show()
```

```python

kwargs = dict(alpha=0.5, bins=5)

plt.hist(kenworth_brian['Preschool Children in Blindzone'], **kwargs, color='g', label='Brian Drawings')
plt.hist(kenworth_eric['Preschool Children in Blindzone'], **kwargs, color='b', label='Eric Drawings')
plt.gca().set(title='Frequency Histogram of Diamond Depths', ylabel='Frequency')
plt.legend();
```
